## **Introduction**

In this project, we explore a fascinating statistical pattern that emerges in language and natural datasets:

**Zipf's Law** – A rule that describes the frequency distribution of words and other ranked items in real-world datasets.

This law reveals hidden structures in seemingly disordered data and finds applications in fields like linguistics, information retrieval, natural language processing (NLP), and data science.

---

## **What is Zipf's Law?**

Zipf’s Law is an empirical law formulated using mathematical statistics that refers to the frequency of words in a language. According to Zipf's Law, in a given body of text, the frequency of any word is **inversely proportional to its rank** in the frequency table.

In simpler terms:

* The most frequent word will appear about **twice as often** as the second most frequent word, **three times as often** as the third most frequent word, and so on.

### **Mathematical Expression:**

If:

* *f(r)* = frequency of the word with rank *r*
  Then:

> **f(r) ∝ 1 / r**

Or formally:

> **f(r) = c / rᵃ**, where *c* is a constant and *a* is usually close to 1.

---

## **Key Takeaways:**

* **Power-law distribution**: Zipf’s Law is a specific case of a power-law distribution, common in many complex systems.
* **Language & beyond**: While most prominent in natural languages, it also appears in population distributions, website visits, income rankings, etc.
* **Scalability**: The law holds across different languages, texts, and dataset sizes (after a certain threshold).

---

## **Examples:**

* **Natural language**: In English, common words like "the", "of", "and" appear far more often than less common words. For example, "the" might appear thousands of times in a novel, while rare words appear only once or twice.
* **Search engine queries**: Some queries are made frequently (e.g., “weather”), while others are asked rarely – but the frequency still follows a Zipf-like distribution.
* **Web traffic**: A few websites receive the majority of internet traffic, while most sites have very few visitors.

---

## **Important Considerations:**

* **Not exact but approximate**: Zipf’s Law isn't always perfect. Deviations often occur, especially in short or structured datasets.
* **Pre-processing matters**: When applying Zipf’s Law to text, removing punctuation, stopwords, and case-folding is essential to obtain accurate frequency distributions.
* **Limitations in interpretation**: While the law reveals patterns, it doesn't always explain why those patterns exist. It's descriptive, not explanatory.

---

## **Conclusion**

Zipf's Law uncovers powerful insights about the structure of language and other ranked data. By understanding and applying this law, we can:

* Identify anomalies in text
* Optimize information retrieval systems
* Build better NLP tools

```python
import pandas as pd

df = pd.read_csv('ColdPlay.csv')
df.head()
```

```python
df.columns

df['Lyric'] = df['Lyric'].astype(str)

all_lyrics = " ".join(df['Lyric'].tolist()).lower()

all_lyrics_clean = re.sub(r'[^a-z\s]', '', all_lyrics)
words = all_lyrics_clean.split()
```

```python
# Frequency count
word_freq = Counter(words)
common_words = word_freq.most_common(30)

# Bar plot
freq_df = pd.DataFrame(common_words, columns=['word', 'count'])
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='word', data=freq_df, hue='word', palette='magma', legend=False)
plt.title("Top 30 Most Common Words in Coldplay Lyrics")
plt.xlabel("Count")
plt.ylabel("Word")
plt.show()


```

```python
# Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Lyrics")
plt.show()
```

```python
# Rank vs Frequency
frequencies = np.array([freq for word, freq in word_freq.most_common()])
ranks = np.arange(1, len(frequencies)+1)

plt.figure(figsize=(10, 6))
plt.loglog(ranks, frequencies, marker=".")
plt.title("Zipf's Law: Word Frequency vs Rank (Log-Log Plot)")
plt.xlabel("Rank of word")
plt.ylabel("Frequency of word")
plt.grid(True)
plt.show()
```

## Final Observations & Insights

- Coldplay's lyrics show **strong adherence to Zipf’s Law**, with the log-log plot forming an almost straight line.
- Frequent words include **emotionally charged or repetitive terms** like “love”, “oh”, and “you”.
